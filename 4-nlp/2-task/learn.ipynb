```python
from textblob import TextBlob
from textblob.np_extractors import ConllExtractor
```

```python
extractor = ConllExtractor()
```

```python
print("How are you today, mate?")
while True:
    message = input("> ")
    print(f"> {message}")

    if message.lower() == "bye":
        print("Okay, have a good day")
        break;

    message_blob = TextBlob(message, np_extractor=extractor)

    if message_blob.polarity <= -0.5:
        response = "Oh dear, that's sound bad"
    elif message_blob.polarity <= 0:
        response = "Hmm, that's not great"
    elif message_blob.polarity <= 0.5:
        response = "That's sound positive"
    else:
        response = "Wow, that's sound great"

    print(f"{response}. Can you tell me more about that things?")
```
